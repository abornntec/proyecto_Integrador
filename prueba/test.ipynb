```python
import pandas as pd
from urllib.parse import urlparse

df = pd.read_csv("URLs.csv")

def extract_lens_key(url: str) -> str:
    path = urlparse(url).path.strip("/")           
    parts = path.split("/")
    dir_name = parts[-2]                           
    key = dir_name
    if key.endswith("-lens"):
        key = key[:-5]
    if key.endswith("-applications"):
        key = key[:-13]
    return key

LENSES = {}
for _, row in df.iterrows():
    url = row["URL"]
    lens_key = extract_lens_key(url)             
    level = None
    title = None
    for lvl_col in ["1st Level","2nd Level","3rd Level","4th Level"]:
        if pd.notna(row[lvl_col]):
            level = int(lvl_col[0])               
            title = row[lvl_col]
    LENSES.setdefault(lens_key, []).append({
        "url":      url,
        "level":    level,
        "section":  title
    })

import pprint
pprint.pprint(LENSES, width=120)
```

```python
import requests
from bs4 import BeautifulSoup
from langchain.schema import Document

def fetch_lens_docs(url: str) -> str:
    resp = requests.get(url)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")
    main = soup.select_one("#content") or soup.body
    return main.get_text(separator="\n", strip=True)

docs = []
for slug, pages in LENSES.items():
    for page in pages:                          
        url     = page["url"]
        level   = page["level"]
        section = page["section"]

        text = fetch_lens_docs(url)
        docs.append(
            Document(
                page_content=text,
                metadata={
                    "lens":    slug,
                    "level":   level,
                    "section": section,
                    "source":  url
                }
            )
        )
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    separators=["\n\n", "\n", " ", ""]
)
chunks = []
for doc in docs:
    for chunk in splitter.split_text(doc.page_content):
        chunks.append(Document(page_content=chunk, metadata=doc.metadata))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

hf_embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
)

index = FAISS.from_documents(chunks, hf_embeddings)

index.save_local("wa_lenses_index_hf")
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain_ollama.llms import OllamaLLM


prompt = """
Eres un clasificador de AWS Well-Architected Lenses.
Dado este fragmento de la documentación (contexto):
{context}

Y la entrada del usuario (consulta):
{question}

Responde SOLO con la clave del lens al que corresponde 
(p.ej. generative-ai, analytics, serverless), sin nada más.
Lens:
"""
template = PromptTemplate(
    template=prompt,
    input_variables=["context", "question"]
)

llm = OllamaLLM(model="gemma3", temperature=0)

qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=index.as_retriever(search_kwargs={"k": 5}),
    return_source_documents=True,
    chain_type_kwargs={"prompt": template}
)

def classify_lens(text: str):
    return qa({"query": text})
```
